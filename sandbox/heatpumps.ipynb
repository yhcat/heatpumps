```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def calculate_cop_by_temperature(T_env, T_target=20):
    # write docstring
    """
    Parameters
    ----------
    T_env : float
        The temperature of the environment in degrees C. default = 0
    T_target : float
        The target temperature of the home in degrees C. default = 20
    """

    # calculate coefficient of performance
    TH = T_target + 273.15
    TE = T_env + 273.15
    cop = TH/(TH - TE)

    return cop

# function to convert temperature to degrees C
def convert_to_celsius(temp):
    # write docstring
    """
    Parameters
    ----------
    temp : float
        The temperature in degrees F.
    """

    # convert temperature to degrees C
    temp_celsius = (temp - 32) * 5/9

    return temp_celsius
```

```python
# generate example T_env from -20 to 18 degrees C
T_env = np.arange(-20, 0, 1)

# calculate COP for each T_env
cop = calculate_cop_by_temperature(T_env)

# plot COP vs. T_env
plt.plot(T_env, cop)
plt.xlabel('T_env (C)')
plt.ylabel('COP')
plt.title('COP vs. T_env')
plt.show()
```
